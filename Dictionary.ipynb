```python
keys=["movie","hero","release year"]
values=["kantara","rishab",2022]
mn={}
for i in range(len(keys)):
  mn[keys[i]]=values[i]
print(mn)
```

```python
mn={}
for i,j in zip(keys,values):
  mn[i]=j
print(mn)
```

```python
mn=dict(zip(keys,values))
print(mn)
```

```python
nums = [1, 2, 3, 4]
squares = []
for n in nums:
  squares.append(n ** 2)
print(nums)
print(squares) #CORRECT
squares = []
for i in range(len(nums)):
   nums[i] = nums[i] ** 2
   squares.append(nums[i])
print(nums)
print(squares)
```
