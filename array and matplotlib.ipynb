```python
#crop yields for 4 farms over 5 years
import numpy as np
crops_yields=np.array([
    [12.5,13.2,14,15,13],
    [13,14.2,15.6,17,12],
    [12.3,13.4,13.2,14.5,14.1],
    [12.3,14.1,15.3,14.3,15],
])

print("crop yirlds array(tons)")
print(crops_yields)
print(f"\nShape: {crops_yields.shape}")
print(f"data type:{crops_yields.dtype}")
```

```python
#Agricultural analysis operations
total_yield=np.sum(crops_yields )
avg_yield_per_farm=np.mean(crops_yields, axis=0)
avg_yield_per_season=np.mean(crops_yields, axis=1)
max_yield=np.max(crops_yields)
min_yield=np.min(crops_yields)

print(f"Total production: {total_yield} tons")
print(f"Average production per farm: {avg_yield_per_farm}tons")
print(f"Average production per season: {avg_yield_per_season}tons")
print(f"maximum yield: {max_yield}tons")
print(f"minimum yield: {min_yield}tons")
```

```python
#best performing farm
best_farm=np.argmax(avg_yield_per_farm)
#np.argmax does not return the max value itself - it returns where the value is located
print("\nBest performing farm:farm",best_farm+1)
print("their average yield", round(avg_yield_per_farm[best_farm],2),"tons")

#or
print(f"\nBest performing farm:farm {best_farm+1}")
print(f"their average yield : {avg_yield_per_farm[best_farm]:.2f} tons")
```

```python
#boolean indexing: find exceptional yields
exceptional=crops_yields>15
print("\n exceptional yield",crops_yields[ exceptional])
print(exceptional)
```

```python
categories=np.where(crops_yields>16,"High",
                 np.where(crops_yields>14,"Medium","Low"))
print(categories)
```
