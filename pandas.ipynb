```python
import pandas as pd
data={
    'farm':["fram 1 ","farm 2","farm 3","farm 4"],
    'crop':["wheat","rice","wheat","barley"],
    'area_in_hectares':[2,3,4,5],
     "productons_in_tons":[30,40,50,60]
}
df=pd.DataFrame(data,index=["a",'b','c','d'])
df["yield_in_tons_per_hectares"]=df["productons_in_tons"]/df["area_in_hectares"]
print(df)
```

```python
df.iloc[[1,3]]
```

```python
df.iloc[[1,3],:3]
```

```python
#filter data
print(df["crop"]=="wheat")
print()
wheat_farms=df[df["crop"]=="wheat"]
print(wheat_farms)
```

```python
sorted_df=df.sort_values(by="area_in_hectares",ascending=False)
print(sorted_df)
```

```python
avg_yield=df.groupby("crop")["yield_in_tons_per_hectares"].mean()
print(avg_yield)
```

```python
tot_yield=df.groupby("crop")["yield_in_tons_per_hectares"].sum()  #wheat gets added
print(tot_yield)
```

```python
#filter data
wheat_data=df[df['Crop']=="Wheat"]
print("wheat_data")
print(wheat_data)
```

```python
#group by year and sum of production
yearly_production=df.groupby("Year")["Production_Tons"].sum()
print(yearly_production)
```

```python
#scatter plot- rainfall vs production
import matplotlib.pyplot as plt

plt.scatter(df["Rainfall_mm"],df["Production_Tons"])
plt.xlabel("Rainfall(mm)")
plt.ylabel("production(tons)")
plt.title("rainfall vs production")
plt.show()
```

```python
#group by year and sum of production
yearly_production=df.groupby("Year")["Production_Tons"].sum()
print(yearly_production)
print(yearly_production.index)
print(yearly_production.values)
```

```python
#create bar plot
plt.figure(figsize=(5,8))
plt.bar(yearly_production.index,yearly_production.values)

plt.xlabel("year")
plt.ylabel("Total production(tons)")
plt.title("Total production by year")

plt.xticks(yearly_production.index)
plt.show()
```

```python
#filter data for each crop
wheat=df[df["Crop"]=="Wheat"]
rice=df[df["Crop"]=="Rice"]

plt.figure()
plt.plot(wheat["Year"],wheat["Production_Tons"],label="wheat")
plt.plot(rice["Year"],rice["Production_Tons"],label="rice")
plt.xlabel("year")
plt.ylabel("production(tons)")
plt.title("production of wheat and rice")
plt.legend()
plt.show()
```

```python
#bar plot - production by crop (single year example: 2022)
#filter year 2022
data_2022=df[df["Year"]==2022]
print(data_2022)

plt.figure(figsize=(4,5))
plt.bar(data_2022["Crop"],data_2022["Production_Tons"],color=["green","orange"])

plt.xlabel("crop")
plt.ylabel("production(tons)")
plt.title("production by crop in 2022")
plt.show()
```

```python
#custom styled bar plot
plt.figure(figsize=(4,5))

bars=plt.bar(yearly_production.index,
             yearly_production.values,
             color="yellow",
             edgecolor="red")

#add values labels on top of bars
for bar in bars:
    height=bar.get_height()
    plt.text(bar.get_x()+bar.get_width()/2,
             height,
             str(height),
             ha="center",
             va="bottom")

plt.xlabel("year")
plt.ylabel("total production(tons)")
plt.title("total production by year")
plt.show()
```
