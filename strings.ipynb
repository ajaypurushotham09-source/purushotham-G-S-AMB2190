```python
import random

# Initialize inventory and money 🎒💰
inventory = {}
total_money = 0

# Fish types and prices 🐟
fish_prices = {
   "Rohu": 12,
    "Catla": 18,
    "Sinchu": 25
}

print("\n🎣 --- Fishing Game Menu --- 🎣")
print("1️⃣ Catch Fish")
print("2️⃣ View Inventory")
print("3️⃣ Sell Fish")
print("4️⃣ Exit 🚪")

# Game loop 🔁
while True:
    choice = int(input("Choose an option: "))

    # Catch Fish 🐠
    if choice == 1:
        fish = random.choice(list(fish_prices.keys()))
        inventory[fish] = inventory.get(fish, 0) + 1
        print(f"🎉 You caught a {fish}!")

    # View Inventory 🎒
    elif choice == 2:
        if inventory == {}:       #if len(inventory) == 0:      #if not inventory:
            print("❌ No fish.")
        else:
            print("📦 Your Inventory:")
            for fish, count in inventory.items():
                print(f"{fish}: {count}")

    # Sell Fish 💵
    elif choice == 3:
        if inventory == {}:
            print("❌ No fish to sell.")
        else:
            earnings = 0
            for fish, count in inventory.items():
                earnings += fish_prices[fish] * count
            total_money += earnings  # total_money = total_money earnings
            inventory.clear()
            print(f"💰 You sold all fish for ${earnings}!")

    # Exit 🚪
    elif choice == 4:
        print(f"🏁 Final money earned: ${total_money}")
        print("👋 Thanks for playing!")
        break

    else:
        print("⚠️ Invalid choice. Try again.")

```
